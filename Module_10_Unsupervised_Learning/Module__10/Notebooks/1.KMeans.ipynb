# K-Means clustering

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("movie_metadata1.csv")
data.head()
```

```python
data.shape
```

```python
list(data.columns)
```

```python
data['director_facebook_likes'].head()
```

```python
new_data = data[['director_facebook_likes', 'actor_3_facebook_likes']]
new_data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# instantiate the model
kmeans = KMeans(n_clusters=5)
```

```python
# train the model
kmeans.fit(new_data)
```

```python
# init='k-means++, the random selection is optimal
kmeans.cluster_centers_
```

```python
# clustering result meaning each row is put in which cluster ids(0,1,2....)
list(kmeans.labels_)
```

```python
import numpy as np
unique, counts = np.unique(kmeans.labels_,return_counts=True)
dict(zip(unique, counts))
```

```python
# plot the data
new_data['cluster'] = kmeans.labels_
import seaborn as sns
sns.lmplot('director_facebook_likes', 'actor_3_facebook_likes', data=new_data,
          hue='cluster', palette='coolwarm', size=6, aspect=1, fit_reg=False)
```
