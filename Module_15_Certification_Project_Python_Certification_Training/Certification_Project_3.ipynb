```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_train='/Users/atul595525/Desktop/learning/Edureka_courses/Python_For_Data_Science/Module_15_Certification_Project_Python_Certification_Training/574_cert_proj_dataset_v3.0/train.csv'

data_test='/Users/atul595525/Desktop/learning/Edureka_courses/Python_For_Data_Science/Module_15_Certification_Project_Python_Certification_Training/574_cert_proj_dataset_v3.0/test.csv'
```

```python
df_test=pd.read_csv(data_test)

df_test.head(2)
```

```python
#2) Import the required classification libraries along with pandas, numpy, seaborn etc
#3) Then import the classifiers from them (Randomforest, SVM, NaiveBayes,
#DecisionTrees)

from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC

```

```python
#4) After this create a function to encode the labels of the strings given in the dataset

df_train.head(2)
```

```python
#4) After this create a function to encode the labels of the strings given in the dataset
#5) You can do the above step using label encoder. With this you are creating some labelsfrom train set as test set. 
#The test set we imported is for testing the best classifier accuracy once we choose it


# Encoding Data
labelencoder = LabelEncoder()
df_train["species"] = labelencoder.fit_transform(df_train["species"])
df_train.head()
```

```python
# Defining dependent and Independent data
X = df_train.iloc[:, 2:]
Y = df_train["species"]

X.head(2)
```

```python
Y.head(2)
```

```python
#6) Then extract the values from train set by stratifying them and dividing it into 80:20
#ratio
#7) Now your X train, X test, Y train, Y test are ready.

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
#8) We currently don’t know which is the best classifier on the dataset. So, we apply all 4 of them.


# 1. Random Forest Classifier
ran_class = RandomForestClassifier(n_estimators=5, random_state=0)
ran_class.fit(x_train, y_train)

predict_rf = ran_class.predict(x_test)
print("Radom Forest Accuracy Score : ",
      metrics.accuracy_score(predict_rf, y_test))

#print(metrics.classification_report(y_test, predict_rf))

```

```python
# 2. Decision Tree Classifier
dec_class = DecisionTreeClassifier(random_state=0)
dec_class.fit(x_train, y_train)

predict_dt = dec_class.predict(x_test)
print("Decision Tree Accuracy Score : ",
      metrics.accuracy_score(predict_dt, y_test))

```

```python
# 3. Naive Bayes Classifier
gn_class = GaussianNB()
gn_class.fit(x_train, y_train)

predict_gn = gn_class.predict(x_test)
print("Naive Bayes Accuracy Score : ",
      metrics.accuracy_score(predict_gn, y_test))

```

```python

# 4. SVM Classifier
svm_class = LinearSVC()
#model = svm.SVC(kernel='linear')
svm_class.fit(x_train, y_train)

predict_svm = svm_class.predict(x_test)
print("SVM Accuracy Score : ", metrics.accuracy_score(predict_svm, y_test))


```

```python
# Best Classifier - SVM Classifier
```
