```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data='/Users/atul595525/Desktop/learning/Edureka_courses/Python_For_Data_Science/Module_14_Model_Selection_and_Boosting/Diabetes.txt'

df=pd.read_csv(data,sep=",",header=None)

df.head(6)
```

```python
df.columns=['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
df.head(8)
```

```python
X=df.iloc[:,0:8]
Y=df.iloc[:,8]
```

```python
Y.head(7)
```

```python
X.head(7)
```

```python
from sklearn.model_selection import KFold
from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier

kfold=KFold(n_splits=10,random_state=7)


model=AdaBoostClassifier(n_estimators=30,random_state=7)

results=model_selection.cross_val_score(model,X,Y,cv=kfold)

print(results.mean())
```
