```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data='/Users/atul595525/Desktop/learning/Edureka_courses/Python_For_Data_Science/Module_15_Certification_Project_Python_Certification_Training/uv2towea1d/OnlineNewsPopularity.csv'

df=pd.read_csv(data)

df.head(6)
```

```python
#dropping timedelta and URL

X=df.iloc[:,2:-1]

X.head(2)
```

```python
#splitting data into trainning and testing set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
#applying linear regression model

from sklearn.linear_model import LinearRegression
from sklearn import metrics
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
#predictions += lin_reg.predict(test_X)


#testing or rpeediction

y_pred = lin_reg.predict(x_test)


```

```python
#Interpreting Coefficients

print(lin_reg.intercept_)

#print the beta coeeficients in same order as passed

print(lin_reg.coef_)


```

```python
#Model Evaluation

#There are three primary metrics used to evaluate linear models. 
'''
These are: 
Mean absolute error (MAE), 
Mean squared error (MSE), or 
Root mean squared error (RMSE).

MAE: The easiest to understand. Represents average error
    
MSE: Similar to MAE but noise is exaggerated and larger errors are “punished”. 
    It is harder to interpret than MAE as it’s not in base units, however, it is generally more popular.

RMSE: Most popular metric, similar to MSE, however, the result is square rooted to make it more 
    interpretable as it’s in base units. It is recommended that RMSE be used as the primary metric to interpret your model.
'''
```

```python
from sklearn import metrics

#Print MAE

print(metrics.mean_absolute_error(y_test,y_pred))

#print result of MSE

print(metrics.mean_squared_error(y_test,y_pred))

#print result of RMSE

print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#Approach 4: Considering time delta also in features

from sklearn import metrics
from matplotlib.pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data1='/Users/atul595525/Desktop/learning/Edureka_courses/Python_For_Data_Science/Module_15_Certification_Project_Python_Certification_Training/uv2towea1d/OnlineNewsPopularity.csv'

#df=pd.read_csv(data)


data = pd.read_csv(data1)
data.head()



# Analysing data & finding dependent and independent variables
X = data.iloc[:, 1:60]
X.head()

corr = X.corr()
sns.heatmap(corr)
rcParams['figure.figsize'] = 20, 20

Y = data["shares"]
Y.head()
```

```python
# Splitting data into training and testing
train_x, test_x, train_y, test_y = train_test_split(
    X, Y, random_state=5, test_size=0.20)


```

```python
# Initializing model and training model
lin_model = LinearRegression()
lin_model.fit(train_x, train_y)


```

```python
# Predict data
predicted = lin_model.predict(test_x)


```

```python
print(np.sqrt(metrics.mean_squared_error(predicted, test_y)))
```
